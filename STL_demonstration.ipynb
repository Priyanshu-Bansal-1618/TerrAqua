</B>Analysis of DWLR data of Chandigarh sector 27</B>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ch=pd.read_csv("chsec27.csv")
```

```python
ch.head(50)
```

```python
print(ch.shape)
```

```python
ch.info()
```

```python

ch = ch[['Data Time', 'Data Value']]
ch.head()
```

```python
ch.shape
```

```python
ch['Data Time'] = pd.to_datetime(ch['Data Time'])

#keeping only 0:00, 6:00,12:00,18:00 values
ch = ch[ch['Data Time'].dt.hour.isin([0, 6, 12, 18])]

print(ch.shape)
print(ch.head())
ch.shape
```

```python
#plotting the raw data
plt.figure(figsize=(12,6))
plt.plot(ch['Data Time'], ch['Data Value'], label="Water Level")
plt.xlabel("Time")
plt.ylabel("Water Level (m)")
plt.title("Groundwater Level Over Time in Chandigarh Sec 27")
plt.legend()
plt.show()
```

```python
#apply seasonal decomposition
from statsmodels.tsa.seasonal import STL
ts=ch['Data Value']
stl = STL(ts, period=4*7)  # weekly seasonal period
res = stl.fit()
fig = res.plot()
fig.set_size_inches(32, 20)
plt.gcf().set_dpi(300)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

residuals = res.resid  # pandas Series
mean = residuals.mean()
std_dev = residuals.std()

# 3-sigma thresholds
upper_threshold = mean +  4* std_dev
lower_threshold = mean - 4 * std_dev

anomaly_indices = np.where((residuals > upper_threshold) | (residuals < lower_threshold))[0]

anomalies = residuals.iloc[anomaly_indices]
time_stamps = ch['Data Time'].iloc[anomaly_indices]

plt.figure(figsize=(15, 6))
plt.scatter(time_stamps, anomalies, color='red', label='3-sigma anomalies')
plt.plot(ch['Data Time'], residuals, alpha=0.5, label='Residuals')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Residual')
plt.title('Residuals with >3 Sigma Deviations Highlighted')
plt.show()
```

```python
#plot the same points breaching threshold on original data
plt.figure(figsize=(12,6))
plt.plot(ch['Data Time'], ch['Data Value'], label="Water Level")
plt.scatter(time_stamps, ch['Data Value'].iloc[anomaly_indices], color='red', label='3-sigma anomalies')
plt.xlabel("Time")
plt.ylabel("Water Level (m)")
plt.title("Groundwater Level Over Time in Chandigarh Sec 27 with Anomalies")
plt.legend()
plt.grid(True)
plt.show()
```
